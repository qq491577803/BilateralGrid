```python
%matplotlib inline

import os
import imageio
import math
from pathlib2 import Path

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.ndimage import gaussian_filter
from scipy.ndimage import convolve1d
from scipy.interpolate import interp2d
```

```python
def gaussian_kernel(x, sig):
    ker = np.exp(-0.5*x**2/sig**2)
    return ker / ker.sum()

class BilateralGrid1D:
    def __init__(self, sam_value: float, sam_space:int, sig_value: float, sig_space: float):
        self.sam_value = sam_value
        self.sam_space = sam_space
        self.sig_value = sig_value
        self.sig_space = sig_space
        
    def make_gaussian_kernel(self, r_value, r_space):        
        v = np.arange(-r_value, (r_value + 1 ), 1) * self.sam_value
        ker_value = gaussian_kernel(v, self.sig_value)
        
        x = np.arange(-r_space, (r_space + 1 ), 1) * self.sam_space
        ker_space = gaussian_kernel(x, self.sig_space)
        
        plt.plot(ker_space)
        
        return ker_value, ker_space

    def make_grid(self, y_norm):
        n = len(y_norm)
        h = math.ceil(1 / self.sam_value)
        w = math.ceil(n / self.sam_space)  #(n + sam_space - 1) // sam_space
        grid  = np.zeros(shape=(2, h, w))

        for i, y in enumerate(y_norm):
            grid[0, math.floor(y / self.sam_value), i // self.sam_space] += y
            grid[1, math.floor(y / self.sam_value), i // self.sam_space] += 1
        return grid

    def proc_grid(self, grid):
        grid_bf = np.empty_like(grid) 
        ker_value, ker_space = self.make_gaussian_kernel(7, 10)
        grid_bf[0] = convolve1d(grid[0], ker_value, axis=0)
        grid_bf[0] = convolve1d(grid[0], ker_space, axis=1)
        grid_bf[1] = convolve1d(grid[1], ker_value, axis=0)
        grid_bf[1] = convolve1d(grid[1], ker_space, axis=1)
#         grid_bf[0] = gaussian_filter(grid[0], sigma=(self.sig_value, self.sig_space))
#         grid_bf[1] = gaussian_filter(grid[1], sigma=(self.sig_value, self.sig_space))
        return grid_bf

    def proc_grid2(self, grid):
        grid_bf = np.empty_like(grid) 
#         ker_value, ker_space = self.make_gaussian_kernel(7, 10)
#         grid_bf[0] = convolve1d(grid[0], ker_value, axis=0)
#         grid_bf[0] = convolve1d(grid[0], ker_space, axis=1)
#         grid_bf[1] = convolve1d(grid[1], ker_value, axis=0)
#         grid_bf[1] = convolve1d(grid[1], ker_space, axis=1)
        grid_bf[0] = gaussian_filter(grid[0], sigma=(self.sig_value, self.sig_space))
        grid_bf[1] = gaussian_filter(grid[1], sigma=(self.sig_value, self.sig_space))
        return grid_bf

    def slice_grid(self, y_norm, grid):
        # bilinear interpolation
        n = len(y_norm)
        h_range = np.arange(0, n, self.sam_space)
        v_range = np.arange(0, 1, self.sam_value)

        f_0 = interp2d(h_range, v_range, grid[0], kind='linear')
        f_1 = interp2d(h_range, v_range, grid[1], kind='linear')

        h_range = np.arange(0, n)
        
        grid_u = np.empty(shape=(2, 32, n))
        grid_u[0] = f_0(h_range, np.linspace(0, 1, 32))
        grid_u[1] = f_1(h_range, np.linspace(0, 1, 32))

        w_bf = np.empty_like(y_norm)
        c_bf = np.empty_like(y_norm)
        for i, y in enumerate(y_norm):
            w_bf[i] = f_0(i, y)
            c_bf[i] = f_1(i, y)
        return w_bf / c_bf, grid_u
    
    def show_grid(self, y, grid_u, ax, show_grid=True):
        n = grid_u.shape[-1]
        h = math.ceil(1 / self.sam_value)
        w = math.ceil(n / self.sam_space)  #(n + sam_space - 1) // sam_space
        h_range = np.arange(0, n, self.sam_space)
        v_range = np.arange(0, 1, self.sam_value) * grid_u.shape[-2]
        if show_grid:
            ax.vlines(h_range, ymin=0, ymax=grid_u.shape[-2], linewidth=1)
            ax.hlines(v_range, xmin=0, xmax=n, linewidth=1)
        ax.imshow(grid_u, cmap='gray', origin="lower")
        ax.plot(y * grid_u.shape[-2], 'r-')
        ax.set_ylim([0, grid_u.shape[-2]])
        ax.set_xlim([0, n])
```

## BG Example

```python
def generate_1d_signal(gain=500):
    np.random.seed(0)
    x = np.linspace(0, 128, 128)
    yy = (0.7 - 0.28) / (30 - 90) * (x-30) + 0.7
    
    y = np.where((30 < x) & (x <= 90), yy, 0.28)
    y[x > 90] = 0.7
    
    sig = np.random.poisson(y * gain)
    sig = sig / gain
    sig = np.clip(sig, 0.0, 1.0)
    return sig
```

```python
bg = BilateralGrid1D(
    sam_value=0.03,
    sig_value=3.5,

    sam_space=8,
    sig_space=1
)

y = generate_1d_signal(gain=500)
grid    = bg.make_grid(y)
grid_bf = bg.proc_grid2(grid)
slc, grid_u  = bg.slice_grid(y, grid_bf)


fig, ((ax1, ax2)) = plt.subplots(nrows=2, sharex='col')
ax1.plot(slc, "r-")
ax1.plot(y, color='gray', ls='-', alpha=0.3)

bg.show_grid(y, grid_u[0], ax2);
```

```python
ltm_curve = np.array([
    [0.000000, 0.000000],
    [0.000028, 0.001537],
    [0.000040, 0.007063],
    [0.000060, 0.017235],
    [0.000080, 0.032454],
    [0.000110, 0.053024],
    [0.000139, 0.079191],
    [0.000182, 0.111162],
    [0.000223, 0.149121],
    [0.000281, 0.193230],
    [0.000334, 0.243636],
    [0.000410, 0.300473],
    [0.000574, 0.363865],
    [0.000962, 0.433928],
    [0.002350, 0.510768],
    [0.011968, 0.594488],
    [0.055835, 0.685182],
    [0.151204, 0.782942],
    [0.324518, 0.887854],
    [1.000000, 1.000000],
])


def simple_tone_map(img):
    curve = ltm_curve.copy()
    s = img
    
    s_interp = np.interp(s, curve[:, 0], curve[:, 1])
    return s_interp
```

## Peak

```python
def construct_1d_peak(gain=500):
    np.random.seed(5)
    x = np.linspace(0, 128, 128)    
    y = np.zeros_like(x)
    
    y =  (np.exp(-(x-50)**2/20)*0.7 + 0.1)
#     y = np.clip(y, 0, 1)

    sig = np.random.poisson(y * gain)
    sig = sig / gain
    sig = np.clip(sig, 0.0, 1.0)
    return sig .astype(np.float64)
```

### Case 1: Halo effect

In the following example, the signal is well smoothed by bilateral grid, but a halo will appear in the final result

```python
bg = BilateralGrid1D(
    sam_value=0.1,
    sig_value=1,
    sam_space=16,
    sig_space=1)

luma = construct_1d_peak()
grid    = bg.make_grid(luma)
grid_bf = bg.proc_grid2(grid)
mask, grid_u  = bg.slice_grid(luma, grid_bf)

# mask = gaussian_filter(mask, sigma=2)

luma_tm = simple_tone_map(luma)
mask_tm = simple_tone_map(mask)

detail_gain = (luma_tm / mask_tm - 1.0) * 0.15 + 1.0
mask_gain = mask_tm / mask
gain = mask_gain * detail_gain

luma_result = luma * gain
luma_result = np.clip(luma_result, 0, 1.0)


fig, ax = plt.subplots()
bg.show_grid(luma, grid_u[0], ax)

# show ltm result
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex='all', figsize=(10, 5))

ax1.plot(luma, label="luma", color='gray')
ax1.plot(mask, label="mask", color='red')

ax2.plot(luma_tm, label = "luma_tm", color='gray')
ax2.plot(mask_tm, label = "mask_tm", color='red')

ax3.plot(gain, label="gain")
ax3.plot(luma, color='gray')

ax4.plot(luma_result, label = "result")

for ax in (ax1, ax2, ax3, ax4):
    ax.legend()
    
for ax in (ax1, ax2):
    ax.set_ylim([0, 1.1])
for ax in (ax3, ax4):
    ax.set_xlabel("space")
ax4.set_ylim([0, 1.1])

ax1.set_title("before tone-mapped")
ax2.set_title("after tone-mapped")
ax3.set_title("gain")
ax4.set_title("result luma")
fig.subplots_adjust(hspace=0.3)
```

### Case 2: No halo but less smooth

```python
bg = BilateralGrid1D(
    sam_value=1/32,
    sig_value=0.1,
    sam_space=16,
    sig_space=2)

luma = construct_1d_peak()
grid    = bg.make_grid(luma)
grid_bf = bg.proc_grid2(grid)
mask, grid_u  = bg.slice_grid(luma, grid_bf)

# mask = gaussian_filter(mask, sigma=2)

luma_tm = simple_tone_map(luma)
mask_tm = simple_tone_map(mask)

detail_gain = (luma_tm / mask_tm - 1.0) * 0.15 + 1.0
mask_gain = mask_tm / mask
gain = mask_gain * detail_gain

luma_result = luma * gain
luma_result = np.clip(luma_result, 0, 1.0)


fig, ax = plt.subplots()
bg.show_grid(luma, grid_u[0], ax, show_grid=False)

# show ltm result
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex='all', figsize=(10, 5))

ax1.plot(luma, label="luma", color='gray')
ax1.plot(mask, label="mask", color='red')

ax2.plot(luma_tm, label = "luma_tm", color='gray')
ax2.plot(mask_tm, label = "mask_tm", color='red')

ax3.plot(gain, label="gain")
ax3.plot(luma, color='gray')

ax4.plot(luma_result, label = "result")

for ax in (ax1, ax2, ax3, ax4):
    ax.legend()
    
for ax in (ax1, ax2):
    ax.set_ylim([0, 1.1])
for ax in (ax3, ax4):
    ax.set_xlabel("space")
ax4.set_ylim([0, 1.1])

ax1.set_title("before tone-mapped")
ax2.set_title("after tone-mapped")
ax3.set_title("gain")
ax4.set_title("result luma")
fig.subplots_adjust(hspace=0.3)
```
